### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento



```python
faturamento_loja1 = loja['Preço'].sum()
print(f'Faturamento total da loja 1 é R$ {faturamento_loja1}')

faturamento_loja2 = loja2['Preço'].sum()
print(f'Faturamento total da loja 2 é R$ {faturamento_loja2}')

faturamento_loja3 = loja3['Preço'].sum()
print(f'Faturamento total da loja 3 é R$ {faturamento_loja3}')

faturamento_loja4 = loja4['Preço'].sum()
print(f'Faturamento total da loja 4 é R$ {faturamento_loja4}')
```

# 2. Vendas por Categoria

```python
import matplotlib.pyplot as plt
vendas_por_categoria_loja1 = loja.groupby('Categoria do Produto')['Preço'].count()
vendas_por_categoria_loja1.plot(kind='bar', figsize=(14, 7), color ='blue')
plt.title('Vendas por Categoria - Loja 1')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--')
fig1 = plt.gcf()
fig1.patch.set_edgecolor('black')
fig1.patch.set_linewidth(2)
plt.show()

vendas_por_categoria_loja2 = loja2.groupby('Categoria do Produto')['Preço'].count()
vendas_por_categoria_loja2.plot(kind='bar', figsize=(14, 7), color ='green')
plt.title('Vendas por Categoria - Loja 2')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--')
fig2 = plt.gcf()
fig2.patch.set_edgecolor('black')
fig2.patch.set_linewidth(2)
plt.show()

vendas_por_categoria_loja3 = loja3.groupby('Categoria do Produto')['Preço'].count()
vendas_por_categoria_loja3.plot(kind='bar', figsize=(14, 7), color ='red')
plt.title('Vendas por Categoria - Loja 3')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--')
fig3 = plt.gcf()
fig3.patch.set_edgecolor('black')
fig3.patch.set_linewidth(2)
plt.show()

vendas_por_categoria_loja4 = loja4.groupby('Categoria do Produto')['Preço'].count()
vendas_por_categoria_loja4.plot(kind='bar', figsize=(14, 7), color ='gray')
plt.title('Vendas por Categoria - Loja 4')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--')
fig4 = plt.gcf()
fig4.patch.set_edgecolor('black')
fig4.patch.set_linewidth(2)
plt.show()
```

# 3. Média de Avaliação das Lojas


```python
avaliacao_loja1 = loja['Avaliação da compra'].mean()
print(f'Média de avaliação da loja 1 é {int(avaliacao_loja1 * 100) / 100.0:.2f} de 5')

avaliacao_loja2 = loja2['Avaliação da compra'].mean()
print(f'Média de avaliação da loja 2 é {int(avaliacao_loja2 * 100) / 100.0:.2f} de 5')

avaliacao_loja3 = loja3['Avaliação da compra'].mean()
print(f'Média de avaliação da loja 3 é {int(avaliacao_loja3 * 100) / 100.0:.2f} de 5')

avaliacao_loja4 = loja4['Avaliação da compra'].mean()
print(f'Média de avaliação da loja 4 é {int(avaliacao_loja4 * 100) / 100.0:.2f} de 5')
```

<h1>EXTRA! Visualização do Desempenho Geográfico</h1>

```python
import folium

center_lat = df_consolidados['Latitude'].mean()
center_lon = df_consolidados['Longitude'].mean()

m = folium.Map(location=[center_lat, center_lon], zoom_start=5)

store_colors = {
    'Loja 1': 'blue',
    'Loja 2': 'green',
    'Loja 3': 'red',
    'Loja 4': 'purple'
}

for index, row in df_consolidados.iterrows():
    bubble_size = row['Faturamento Total'] / 50000
    fill_color = store_colors.get(row['Loja'], 'gray')
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=bubble_size,
        color='black',
        weight=1,
        fill_color=fill_color,
        fill_opacity=0.7,
        popup=f"""
        <b>Loja:</b> {row['Loja']}<br>
        <b>Faturamento Total:</b> R$ {row['Faturamento Total']:.2f}<br>
        <b>Média de Avaliação:</b> {row['Média de Avaliação']:.2f}<br>
        <b>Média de Frete:</b> R$ {row['Média de Frete']:.2f}
        """
    ).add_to(m)
display(m)
```
